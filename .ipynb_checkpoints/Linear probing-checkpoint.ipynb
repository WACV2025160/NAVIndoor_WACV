```python
import numpy as np
import torch
import pickle
from tqdm import tqdm
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.metrics import roc_curve, auc, accuracy_score, f1_score
import matplotlib.pyplot as plt

n_scenes = 13
np.random.seed(42)
torch.manual_seed(42)
np.random
train = np.random.choice(13,10,False)
```

# Reproducibility : Due to memory limitations Kernel must be reinitialized for each model. Model features were already computed and saved.

```python



folders = {0:'ConvNextv2_features',
           1:'dinov2_features',
           2:'EfficientNetb7_features',
           3:'segformer_features',
           4:'RL1latent_space',
           5:'RL2latent_space',
           6:'RLfloorlatent_space',
           7:'RLnvlatent_space',
           8:'RL1q_values',
           9:'RL2q_values',
           10:'RLfloorq_values',
           11:'RLnvq_values'
          }
folder = folders[4]


X = []
Y = []
X_test = []
Y_test = []
for n in tqdm(range(n_scenes)):
    with open('distances/'+str(n)+'.pickle', 'rb') as f:
            distances = pickle.load(f)
    with open(folder+'/'+str(n)+'.pickle', 'rb') as f:
            features = pickle.load(f)
  
    if n in train:        
        X.append(features)
        Y.extend(distances)
    else:
        X_test.append(features)
        Y_test.extend(distances)


X_train = torch.concat(X).to('cuda')
Y_train_dist = torch.tensor(Y,dtype=torch.float).to('cuda').unsqueeze(1)
X_test = torch.concat(X_test).to('cuda')
Y_test_dist = torch.tensor(Y_test,dtype=torch.float).to('cuda').unsqueeze(1)
```

```python
t = 3
aucs = []
F1s = []
for t in range(6):
    
    Y_train = (Y_train_dist>t).float()
    Y_test = (Y_test_dist>t).float()
    
    # Create a dataset and loader for both training and testing
    batch_size = 256  # Adjust based on your computational resources
    train_dataset = TensorDataset(X_train, Y_train)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    
    test_dataset = TensorDataset(X_test, Y_test)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    # Define the simple linear probing model
    class LinearProbeNN(nn.Module):
        def __init__(self, num_features, num_classes):
            super(LinearProbeNN, self).__init__()
            self.linear = nn.Linear(num_features, num_classes)  # Single linear layer
    
        def forward(self, x):
            return self.linear(x)
    
    # Initialize the model
    num_features = X_train.shape[1]
    num_classes = 1  # Adjust based on your number of classes
    model = LinearProbeNN(num_features, num_classes).to('cuda')
    
    # Loss and optimizer
    criterion = nn.BCEWithLogitsLoss()  # Suitable for classification tasks
    optimizer = optim.Adam(model.parameters(), lr=0.001)  # Learning rate can be adjusted
    
    # Training loop
    num_epochs = 80 # Adjust based on your needs
    for epoch in range(num_epochs):
        for inputs, labels in train_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
    
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            del inputs
            del labels
            torch.cuda.empty_cache()
    
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
    
    model.eval()
    y_pred = []
    y_prob = []
    with torch.no_grad():
        for inputs, _ in test_loader:
            outputs = model(inputs)
            probs = torch.sigmoid(outputs).cpu().detach()  # Probability of class 1
            #y_pred.extend(preds.numpy())
            y_prob.extend(probs)
    
    y_prob = np.array(torch.stack(y_prob))
    y_pred = y_prob>0.5
    
    fpr, tpr, thresholds = roc_curve(Y_test.cpu().detach(), y_prob)
    roc_auc = auc(fpr, tpr)
    f1 = f1_score(Y_test.cpu().detach(), y_pred)
    # Visualize the ROC curve
    print('FOR DISTANCE ', t, '  F1 SCORE : ', f1, '  AUC : ', roc_auc)
    aucs.append(roc_auc)
    F1s.append(f1)
    del model
    torch.cuda.empty_cache()
```
