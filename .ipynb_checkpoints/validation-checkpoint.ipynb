```python
import scipy.io
from matplotlib import pyplot as plt
import json
from transformers import SegformerImageProcessor, SegformerForSemanticSegmentation

from PIL import Image
import os
from tqdm import tqdm
import torch.nn.functional as F
import torch
import glob
from torch import nn
import torch.nn.init as init
import cv2
import numpy as np
import pandas as pd
import pickle
```

```python
device = 'cuda'
# Assuming you have a processor that can convert logits to predictions
# (e.g., converting logits to segmentation mask using argmax)
processor = SegformerImageProcessor.from_pretrained("nvidia/segformer-b2-finetuned-ade-512-512")
model = SegformerForSemanticSegmentation.from_pretrained("nvidia/segformer-b2-finetuned-ade-512-512").to(device)
```

# Utils

```python


def extract_directions(path):
    image_structs = scipy.io.loadmat(path)
    image_structs = image_structs['image_structs']
    image_structs = image_structs[0]
    
    directions = {}
    centroids = 1/12*np.arange(13) #normalized directions centroids (1 = 2pi)
    
    for camera in image_structs:
        name = camera[0][0]
        camera[4] /= 2;
        theta = np.arctan2(camera[4][2][0], camera[4][0][0])
        if theta<0:
            theta = (np.pi - theta)
        theta/=2*np.pi
        
        direction = np.argmin(np.abs((theta-centroids)),axis=0)
        directions[name] = direction
    return(directions)



def load_scene(path='ActiveVisionDataset',i=0,floor_val=0):
    scenes = os.listdir(path)
    scene_path = path+'/'+scenes[i]
    
    with open(scene_path + '/'+'annotations.json') as json_file:
        data = json.load(json_file)

    images = glob.glob(scene_path + '/'+'jpg_rgb/*')
    image_names = os.listdir(scene_path+'/'+'jpg_rgb/')
    struct_file = scene_path + '/'+'image_structs.mat'
    directions = extract_directions(struct_file)
    images_dict = {}

    for i_path,i_name in zip(images,tqdm(image_names)):
        image = Image.fromarray(np.array(Image.open(i_path))[:,320:-320]).resize((512,512))
        #input = processor(image)['pixel_values']
        images_dict[i_name] = {}
        images_dict[i_name]['img'] = image

        t_floor = 0.2
        t_wall = 0.2
        t_obstacles = 0.3
        
        seg_map = process_logits(process_image(image,model),t_floor = t_floor, t_wall=t_wall,t_obstacles=t_obstacles,floor_val=0)
        seg_map_floor = process_logits(process_image(image,model),t_floor = t_floor, t_wall=t_wall,t_obstacles=t_obstacles,floor_val=0.5)
        images_dict[i_name]['seg'] = seg_map
        images_dict[i_name]['seg_floor'] = seg_map_floor
        images_dict[i_name]['direction'] = directions[i_name]

        #compute distance to endpoint
        k=0
        if data[i_name]['forward']:
            t = i_name
            while data[t]['forward']:
                t = data[t]['forward']
                k+=1
        images_dict[i_name]['distance'] = k
    return(data, images_dict)



#data, image_dict_all = load_scene(i=0,floor_val=0.5)
```

```python



def process_image(img_input,model):
    inputs = processor(images=img_input, return_tensors="pt")
    inputs['pixel_values'] = inputs['pixel_values'].to(device)
    outputs = model(**inputs)
    logits = outputs.logits  # shape (batch_size, num_labels, height/4, width/4)
    logits = logits.transpose(1,3).transpose(1,2) #transpose logits dim to last
    probabilities = F.softmax(logits,dim=-1) #softmax for probas
    logits = probabilities[0].cpu().detach() #get item
    return(logits)


def process_logits(logits, t_floor = 0.2, t_wall=0.2,t_obstacles=0.6,floor_val=0):
    floor = torch.tensor((np.array(logits[:,:,3]>t_floor))*1)
    wall = torch.tensor((np.array(logits[:,:,0]>t_wall))*1)
    obstacles = torch.tensor(((np.array(logits[:,:,oii]).max(axis=-1)>t_obstacles)*1))
    id_floor = torch.where(floor)
    id_wall = torch.where(wall)
    id_obstacles = torch.where(obstacles)
    
    tmp = torch.zeros_like(floor).float()-0.5 #-0.5 : everything else is wall
    tmp[id_floor] = floor_val
    tmp[id_obstacles] = -1
    tmp[id_wall] = -0.5
    return(tmp)

def toint(img):
    return((np.array((img+1)*127.5).astype(np.uint8)))

OI = {
  #"0": "wall",
  #"3": "floor",
  #"8": "windowpane",
  #"14": "door",
  "15": "table",
  "19": "chair",
  "23": "sofa",
  #"26": "sea",
  "30": "armchair",
  "31": "seat",
  "33": "desk",
  "38": "railing",
  "42": "column",
  "64": "coffee table",
  "65": "toilet",
  "69": "bench",
  "75": "swivel chair",
  "77": "bar",
  "95": "bannister",
  "110": "stool",
  "111": "barrel",
  "115": "bag",
  "121": "step",
  "130": "screen",
  "132": "sculpture",
  "147": "glass",
}

oii = torch.tensor([int(i) for i in list(OI.keys())])


class MNmodelPartAM(nn.Module): # A modified version for feature extraction
    def __init__(self,scale,output = 128,dropout=0.25,n_actions=5,am=10,aes=10,ahes=100): #action memory, action embed size, action historic embed size
        super(MNmodelPartAM, self).__init__()
        # Input shape: [batch_size, 3, 64, 64]
        self.unet = False
        self.y_size = 10*am
        self.n_actions = n_actions
        self.am = am
        # Convolutional layers
        self.dropout = dropout
        self.output = output
        self.conv0 = nn.Conv2d(3, 16*scale, kernel_size=5, stride=1,padding=1) #CHECK HOW SPATIAL DIM IS REDUCED AFTER THIS
        self.conv1 = nn.Conv2d(16*scale, 16*scale, kernel_size=3, stride=2,padding=1) #CHECK HOW SPATIAL DIM IS REDUCED AFTER THIS
        self.conv2 = nn.Conv2d(16*scale, 16*scale, kernel_size=3, stride=2,padding=1)
        self.conv3 = nn.Conv2d(16*scale, 4*scale, kernel_size=3, stride=1,padding=1)
        self.conv4 = nn.Conv2d(4*scale, 1*scale, kernel_size=3, stride=2,padding=1)
        # Fully connected layers
        # Calculate the flattened size after the convolutional layers
        self.fc1 = nn.Linear(1*scale*16*16, 1*scale*16*16)
        self.dropout1 = nn.Dropout(self.dropout)
        self.fc2 = nn.Linear(1*scale*16*16, scale*8*16)
        self.dropout2 = nn.Dropout(self.dropout)
        self.fc3 = nn.Linear(scale*8*16, 16*16)
        self.dropout3 = nn.Dropout(self.dropout)
        self.fc4 = nn.Linear(16*16+ahes, 16*16+ahes) #FC4 concat y
        self.dropout4 = nn.Dropout(self.dropout)
        self.fc5 = nn.Linear(16*16+ahes, 16*16+ahes)
        self.dropout5 = nn.Dropout(self.dropout)
        self.fc6 = nn.Linear(16*16+ahes, self.output)
        
        self.fc1_y = nn.Linear(self.n_actions, aes)
        self.fc2_y = nn.Linear(aes*am, ahes)
        self.dropout6 = nn.Dropout(self.dropout)
        self.dropout7 = nn.Dropout(self.dropout)
        
        self.initialize_weights()
    def initialize_weights(self):
        for param in self.parameters():
            if param.requires_grad:
                if len(param.shape) > 1:
                    init.xavier_uniform_(param)
                else:
                    init.constant_(param, 0)
    def forward(self, x, y, eval = False):
        # Apply convolutional layers with GeLU activation and batch normalization
        x = F.gelu(self.conv0(x))
        x = F.gelu(self.conv1(x))
        x = F.gelu(self.conv2(x))
        x = F.gelu(self.conv3(x))
        x = F.gelu(self.conv4(x))
        # Flatten the output for the fully connected layers
        x = x.flatten(1)
        latent = x.clone().cpu().detach()

        # Apply fully connected layers with GeLU activation and dropout (except the last layer)
        x = F.gelu(self.fc1(x))
        if not eval:
            x = self.dropout1(x)
        x = F.gelu(self.fc2(x))
        if not eval:
            x = self.dropout2(x)
        x = F.gelu(self.fc3(x))
        if not eval:
            x = self.dropout3(x)

        latent_2 = x.clone().cpu().detach()
        y = F.gelu(self.fc1_y(y)) #action_mapping
        if not eval:
            y = self.dropout6(y)
        y = y.flatten(-2) #concatenate action embeds
        y = F.gelu(self.fc2_y(y))
        if not eval:
            y = self.dropout7(y)

        x = torch.cat((x,y),axis=-1)
        
        x = F.gelu(self.fc4(x))
        x = F.gelu(self.fc5(x))
        x = self.fc6(x)
        return [latent,latent_2], x

class MNmodelPartAM_Dueling(nn.Module):
    def __init__(self,scale,output = 128,dropout=0.25,n_actions=5,am=10,aes=10,ahes=100,n_frames=3): #action memory, action embed size, action historic embed size
        super(MNmodelPartAM_Dueling, self).__init__()
        # Input shape: [batch_size, 3, 64, 64]
        self.unet = False
        self.y_size = 10*am
        self.n_actions = n_actions
        self.am = am
        # Convolutional layers
        self.dropout = dropout
        self.output = output
        self.conv0 = nn.Conv2d(n_frames, 16*scale, kernel_size=5, stride=1,padding=1) #CHECK HOW SPATIAL DIM IS REDUCED AFTER THIS
        self.conv1 = nn.Conv2d(16*scale, 16*scale, kernel_size=3, stride=2,padding=1) #CHECK HOW SPATIAL DIM IS REDUCED AFTER THIS
        self.conv2 = nn.Conv2d(16*scale, 16*scale, kernel_size=3, stride=2,padding=1)
        self.conv3 = nn.Conv2d(16*scale, 4*scale, kernel_size=3, stride=1,padding=1)
        self.conv4 = nn.Conv2d(4*scale, 1*scale, kernel_size=3, stride=2,padding=1)
        # Fully connected layers
        # Calculate the flattened size after the convolutional layers
        self.fc1 = nn.Linear(1*scale*16*16, 1*scale*16*16)
        self.dropout1 = nn.Dropout(self.dropout)
        self.fc2 = nn.Linear(1*scale*16*16, scale*8*16)
        self.dropout2 = nn.Dropout(self.dropout)
        self.fc3 = nn.Linear(scale*8*16, 16*16)
        self.dropout3 = nn.Dropout(self.dropout)
        self.fc4 = nn.Linear(16*16+ahes, 16*16+ahes) #FC4 concat y
        self.dropout4 = nn.Dropout(self.dropout)

        self.fcv1 = nn.Linear(16*16+ahes, 16*16+ahes)
        self.fcv2 = nn.Linear(16*16+ahes, 100)
        self.fcv3 = nn.Linear(100, 1)
        
        self.fc5 = nn.Linear(16*16+ahes, 16*16+ahes)
        self.dropout5 = nn.Dropout(self.dropout)
        self.fc6 = nn.Linear(16*16+ahes, self.output)
        
        self.fc1_y = nn.Linear(self.n_actions, aes)
        self.fc2_y = nn.Linear(aes*am, ahes)
        self.dropout6 = nn.Dropout(self.dropout)
        self.dropout7 = nn.Dropout(self.dropout)
        
        self.initialize_weights()
    def initialize_weights(self):
        for param in self.parameters():
            if param.requires_grad:
                if len(param.shape) > 1:
                    init.xavier_uniform_(param)
                else:
                    init.constant_(param, 0)
    def forward(self, x, y, eval = False):
        # Apply convolutional layers with GeLU activation and batch normalization
        x = F.gelu(self.conv0(x))
        x = F.gelu(self.conv1(x))
        x = F.gelu(self.conv2(x))
        x = F.gelu(self.conv3(x))
        x = F.gelu(self.conv4(x))
        # Flatten the output for the fully connected layers
        x = x.flatten(1)
        latent = x.clone().cpu().detach()

        # Apply fully connected layers with GeLU activation and dropout (except the last layer)
        x = F.gelu(self.fc1(x))
        if not eval:
            x = self.dropout1(x)
        x = F.gelu(self.fc2(x))
        if not eval:
            x = self.dropout2(x)
        x = F.gelu(self.fc3(x))
        if not eval:
            x = self.dropout3(x)


        y = F.gelu(self.fc1_y(y)) #action_mapping
        if not eval:
            y = self.dropout6(y)
        y = y.flatten(-2) #concatenate action embeds
        y = F.gelu(self.fc2_y(y))
        if not eval:
            y = self.dropout7(y)

        x = torch.cat((x,y),axis=-1)
        
        x = F.gelu(self.fc4(x))
        latent_2 = x.clone().cpu().detach()
        v = F.gelu(self.fcv3(F.gelu(self.fcv2(F.gelu(self.fcv1(x))))))
        
        x = F.gelu(self.fc5(x))
        a = F.gelu(self.fc6(x))
        norm_cst = a.mean()
        
        return [latent,latent_2],v+a-norm_cst



def visualize_arrows_bis(frame, q_value,t=3):
    height, width, _ = frame.shape
    frame = frame.copy()
    if q_value < t:
        color = (255,0,0)
    else:
        color = (0,255,0)
    q_value = min(q_value,6)
    q_value/=10
    # Arrow length proportional to the image size
    arrow_length = min(height, width) // 4

    # Arrow colors (assuming probabilities are in the range [0, 1])

    # Convert angles to radians
    angle = np.radians(90)

        # Calculate arrow tip coordinates


    tip_x = int(width / 4 + arrow_length * q_value * np.cos(angle))
    tip_y = int(height / 4 - arrow_length * q_value * np.sin(angle))

    # Draw arrow
    cv2.arrowedLine(frame, (width // 4, height // 4), (tip_x, tip_y), color, 2)

    return frame
```

```python
def filter_dict(image_dict, direction=0):
    if direction is None:
        return(image_dict)
    filtered_dict = {outer_key: inner_dict for outer_key, inner_dict in image_dict.items() if inner_dict.get('direction', 0) == direction}
    return(filtered_dict)
def extract_values(image_dict,scene_number,results, RL_model,model_name,input_type='seg'):
    results[model_name][scene_number] = {}
    for img_name in tqdm(list(image_dict.keys())):
        
        results[model_name][scene_number][img_name] = {}
        seg = image_dict[img_name][input_type]
        state = [seg,seg,seg]
        action_states = []
        for i in range(20):
            act = torch.zeros(4)
            #act[0] = 1
            action_states.append(act)
        action_input = torch.stack(action_states).unsqueeze(0)
        img_input = torch.stack(state).unsqueeze(0)
        with torch.no_grad():
            [latent,latent_2],q_values = RL_model(img_input,action_input)
        q_values = q_values[0]

        #image_dict['turn_value'] = max(q_values[1],q_values[2]).item()
        results[model_name][scene_number][img_name]['q_values'] = q_values
        #image_dict['q_value_forward'] = q_values[0].item()
        results[model_name][scene_number][img_name]['latent_space'] = latent[0]
        results[model_name][scene_number][img_name]['latent_space_2'] = latent_2[0]
    return(results)
```

# $R_{\theta}(s)$ with respect to d

```python


model_name = 'RL1'
plt.figure(figsize=(8,6))
n_scenes = 13
plt.rcParams['text.usetex'] = False

tmp = {} #will store each x,y for individual plots
tmp_2 = {} #will store the mean_values for each distance as list (for mean plot)
for i in range(n_scenes):
    tmp[i] = []


for n in range(n_scenes):
    print(n)
    with open('scene_'+str(n)+'.pickle', 'rb') as f:
            data,image_dict = pickle.load(f)
    distances = np.array([image_dict[i]['distance'] for i in image_dict.keys()])
    q_values = torch.stack([results[model_name][n][i]['q_values'] for i in results[model_name][n].keys()]).cpu().detach()
    turn_values = q_values.T[1:3].max(axis=0).values
    d, counts = np.unique(distances,return_counts=True)
    d_list = d[np.where(counts>20)[0]] #we remove distances for which very few samples are available (<20)
    mqv = []

    for i in d_list:
        mean_value = ((turn_values-q_values.T[0])[np.array(distances)==i]).mean()
        mqv.append(mean_value)
        if i not in tmp_2.keys():
            tmp_2[i] = []
        tmp_2[i].append(mean_value)
    tmp[n].append([d_list,mqv])

for i in tmp_2.keys():
    if len(tmp_2[i])>0:
        tmp_2[i] = np.array(tmp_2[i]).mean()

for i in tmp.keys():
    d_list, mqv = tmp[i][0]
    plt.plot(d_list,mqv,'r--',c='gray',linewidth=0.3)






x = np.array(list(tmp_2.keys()))
y = np.array(list(tmp_2.values()))

sort = np.argsort(x)
x = x[sort]
y = y[sort]

plt.plot(x,y,'r',c='purple', marker='o')
plt.xlabel('Forward distance to navigation boundary $d$',fontsize=16)
plt.ylabel('$Q_{\\theta}(s,rotate)-Q_{\\theta}(s,forward)$',fontsize=18)
plt.xticks(fontsize=18)  # Change font size of x-axis tick labels
plt.yticks(fontsize=14) 

```

# Our target d correlates with mean depth in image

```python
#Correlation with depth

def compute_depth(path):
    try:
        depth = np.array(Image.fromarray(np.array(Image.open(path))[:,320:-320]).resize((512,512)))
    except:
        return(1000) #very few images semm corrupted, we impute them with value 1000
    mean_depth = depth[depth>0].mean() #removing unknown values
    return(mean_depth.item())


model_name = 'RL1'
plt.figure(figsize=(8,6))
n_scenes = 13

tmp = {} #will store each x,y for individual plots
tmp_2 = {} #will store the mean_values for each distance as list (for mean plot)
for i in range(n_scenes):
    tmp[i] = []

path = 'ActiveVisionDataset'   
scenes = os.listdir(path)

for n in range(n_scenes):
    print(n)
    with open('scene_'+str(n)+'.pickle', 'rb') as f:
            data,image_dict = pickle.load(f)
    distances = np.array([image_dict[i]['distance'] for i in image_dict.keys()])
    scene_path = path+'/'+scenes[n]+'/high_res_depth/'
    inames = [scene_path+i.replace('01.jpg','03.png') for i in list(image_dict.keys())] #03 suffix = depth, 01 for rgb images
    depth = np.array([compute_depth(i) for i in tqdm(inames)])
    
    d, counts = np.unique(distances,return_counts=True)
    d_list = d[np.where(counts>20)[0]]
    mqv = []

    for i in d_list:
        mean_value = depth[np.array(distances)==i].mean()
        mqv.append(mean_value)
        if i not in tmp_2.keys():
            tmp_2[i] = []
        tmp_2[i].append(mean_value)
    tmp[n].append([d_list,mqv])



for i in tmp_2.keys():
    if len(tmp_2[i])>0:
        tmp_2[i] = np.array(tmp_2[i]).mean()

for i in tmp.keys():
    d_list, mqv = tmp[i][0]
    plt.plot(d_list,mqv,'r--',c='gray',linewidth=0.3)






x = np.array(list(tmp_2.keys()))
y = np.array(list(tmp_2.values()))

sort = np.argsort(x)
x = x[sort]
y = y[sort]

plt.plot(x,y,'r',c='purple', marker='o')
plt.xlabel('Forward distance to navigation boundary $d$',fontsize=16)
plt.ylabel('Mean distance (mm)',fontsize=18) #mm (not sure)
plt.xticks(fontsize=18)  # Change font size of x-axis tick labels
plt.yticks(fontsize=14) 
```

# LDA

```python
from tqdm import tqdm
import numpy as np
import pandas as pd
import torch
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, accuracy_score, f1_score
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix# Assuming x is a list of text documents and y is a binary label (0 or 1)
# You need to convert text data into numerical features using CountVectorizer
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
with open('features_RL.pickle', 'rb') as f:
            results = pickle.load(f)
```

```python



distances_Y = []
q_values_X = []
n_scenes = 13

distances_Y_test = []
q_values_X_test = []
train = np.random.choice(12,10,replace=False)
model_name = 'RLfloor'
for n in range(n_scenes):
    with open('scene_'+str(n)+'.pickle', 'rb') as f:
            data,image_dict = pickle.load(f)
    distances = np.array([image_dict[i]['distance'] for i in image_dict.keys()])
    q_values = torch.stack([results[model_name][n][i]['latent_space'] for i in results[model_name][n].keys()]).cpu().detach()
    if n in train:
        q_values_X.extend((q_values.tolist()))
        distances_Y.extend(distances)
    else:
        q_values_X_test.extend((q_values.tolist()))
        distances_Y_test.extend(distances)

t_dist = 3

Y = np.array(distances_Y).flatten()
Y_reg = Y
Y = Y >t_dist
X = np.array(q_values_X)

Y_test = np.array(distances_Y_test).flatten()
Y_reg_test = Y_test
Y_test = Y_test >t_dist
X_test = np.array(q_values_X_test)



# Split the data into training and testing sets
#lda = QuadraticDiscriminantAnalysis() #
lda = LinearDiscriminantAnalysis()
lda = LogisticRegression(max_iter=1000)
lda.fit(X,Y)

# Predict probabilities for the positive class
y_prob = lda.predict_proba(X_test)[:, 1]
y_pred = lda.predict(X_test)

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(Y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Visualize the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for LDA')
plt.legend(loc='lower right')
plt.show()
print('F1 score : ', f1_score(Y_test, y_pred))
```
